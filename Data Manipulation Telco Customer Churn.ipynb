**DATA MANIPULATION**

In this notebook, I perform exercise of data manipulation for `telco_customer_churn.csv` dataset. Data Manipulation aims to adjust the data to become  more understandable for generating analysis

```python
import pandas as pd
import numpy as np
```

After import the libraries needed, i need to read the data set using pandas. The source of dataset comes from kaggle https://www.kaggle.com/datasets/blastchar/telco-customer-churn

```python
df = pd.read_csv('telco_customer_churn.csv')
```

**1. Remove rows with `tenure` = 0 , how many rows are currently available in the data ?**

```python
df = df[df['tenure']!=0].reset_index(drop=True)
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
df.shape
```

The rows are currently available are 7032 rows

**2. What is the average of `tenure` for customer who have no partner and no dependents ?**

answer : 23.303354

**3. What is the maximum `MonthlyCharges` recorded for customer who have partner and no dependents ?**

answer: 11860

**4. Count the customers who have dependents but no partner !**

answer : 359

```python
#Code for answering number 2-4
```

```python
df.groupby(['Partner', 'Dependents']).agg({'tenure':['mean', 'median'],
                                           'MonthlyCharges':['max', 'min'],
                                           'customerID':'count'})
```

**5. What is the average of `TotalCharges` from non-senior citizen customers (Senior citizen = 0) who have both `PhoneServices` and `Internet Sercive Fiber Optic`**

answer : 3226.492737

```python
df.pivot_table(index = ['SeniorCitizen', 'PhoneService'],
              columns = ['InternetService'],
              values = ['TotalCharges'], aggfunc = 'mean')
```
